стэш - временное сохранение изменений, которые не фиксируются в истории проекта
коммит - окончательное сохранение изменений с записью в историю репозитория

  стэш когда нужно сохранить изменения, но не готовы коммитить
  коммит когда нужно зафиксировать изменения в проекте
